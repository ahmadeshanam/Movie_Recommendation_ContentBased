```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("imdb_top_1000.csv")
```

```python
df.head()
```

```python
df.drop(['Poster_Link', 'Runtime','Certificate', 'Overview', 'Released_Year','No_of_Votes'],axis=1,inplace=True)
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
def fill_na_with_imdb(row):
    if pd.isna(row['Meta_score']):
        return row['IMDB_Rating'] * 10
    else:
        return row['Meta_score']

df['Meta_score'] = df.apply(fill_na_with_imdb, axis=1)
```

```python
df['Gross'].fillna(0)
df['Gross'] = df['Gross'].str.replace(',', '')
df['Gross'] = df['Gross'].astype(float)
gross_mean = df['Gross'].mean()
df['Gross'] = df['Gross'].fillna(gross_mean)
df['Gross'].astype(int)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df["Act_Drct"] = df['Star1'].astype(str) + ", " + df['Star2'].astype(str)+ ', ' + df["Star3"].astype(str)+ ', ' + df["Star4"].astype(str) + ', ' + df['Director'].astype(str)
df.drop(['Star1','Star2','Star3','Star4','Director'], axis=1, inplace=True)
df.head()
```

```python
genres = df['Genre'].str.get_dummies(', ')
df = pd.concat([df, genres], axis=1)
df.head()
```

```python
Act_Drct = df['Act_Drct'].str.get_dummies(', ')
df = pd.concat([df, Act_Drct], axis=1)
df.drop(['Genre', 'Act_Drct'],axis=1, inplace=True)
df.head()
```

```python
sample_user_fav = pd.DataFrame() 
sample_user_fav = df[df['Series_Title'].isin(['Whiplash' , 'CODA', 'The Legend of 1900' , 'Inception' , 'Léon', 'The Pianist', 'The Platform','Parasite', 'The Prestige','Interstellar', 'Midnight in Paris', 'Fight Club', 'Inside Out', 'Frozen', 'Elemental', 'Spirited Away', 'Shutter Island', 'Soul', 'Luca', 'Your Name'])]
sample_user_fav
```

```python
sample_user_fav['Rating'] = [9,7.9,9.4,8.4,9.2,6,8,8.6,9.5,7.3,5.8]
print('sample user fav movies and ratings:\n\n',sample_user_fav[['Series_Title','Rating']])
```

```python
user_rates = sample_user_fav.drop(['IMDB_Rating','Meta_score','Gross','Series_Title'],axis=1)
user_rates.head()
```

```python
user_profile = user_rates.transpose().dot(sample_user_fav['Rating'])
user_profile
```

```python
ndf = df.drop(['IMDB_Rating','Meta_score','Gross','Series_Title'],axis=1)
ndf.head()
```

```python
recommendationTable = ((ndf*user_profile).sum(axis=1))/(user_profile.sum())
recommendationTable.head()
```

```python
recommendationTable = recommendationTable.sort_values(ascending=False)
recommendationTable.head()
```

```python
recommendation = df.loc[ndf.index.isin(recommendationTable.head(100).keys())]
```

```python
recommendation = recommendation[~recommendation['Series_Title'].isin(sample_user_fav['Series_Title'])]
```

```python
recommendation.head(20)
```

```python
df_raw = pd.read_csv('imdb_top_1000.csv')
```

```python
recommendation = df_raw[df_raw['Series_Title'].isin(recommendation['Series_Title'])]
```

```python
recommendation.drop(['Certificate','Genre','Star4','Meta_score','No_of_Votes','Gross'],axis=1,inplace=True)
```

```python
recommendation["Actors"] = recommendation['Star1'].astype(str) + ", " + recommendation['Star2'].astype(str)+ ', ' + recommendation["Star3"].astype(str)
recommendation.drop(['Star1','Star2','Star3'],axis=1, inplace=True)
recommendation.rename(columns = {'Poster_Link':'Poster','Series_Title':'Title','IMDB_Rating':'IMDB'}, inplace = True)
```

```python
recommendation.reset_index(inplace=True,drop=True)
print('movies recommended for sample user:\n\n', recommendation[['Title','IMDB']].head(15))
```
